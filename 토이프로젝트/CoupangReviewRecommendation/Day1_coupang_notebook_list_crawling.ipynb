## 크롤링 해볼 쿠팡 링크
- https://www.coupang.com/np/categories/497135

- 1) 이미지, 링크, 상품이름, 가격 추출 
- 2) 해당 링크로 들어가서 리뷰를 추출

```python
import requests
```

```python
# headers는 필수값이 아니지만 쿠팡의 경우 headers 없이 크롤링 했을 땐 결과가 나오지 않음
# headers의 user-agent 정보를 알려줌으로써 봇이 아니라 사람임을 알려주는 것이다
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}

def getPageString(url):
    data = requests.get(url, headers = headers)
    data_content = data.content.decode('utf-8')
    #print(data_content)
    return data_content
```

```python
url = "https://www.coupang.com/np/categories/497135"

# html 통채로 불러와졌다
print(getPageString(url))
```

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup

def getProducts(string):
    rows = []
    bsObj = BeautifulSoup(string, "html.parser")
    ul = bsObj.find("ul", {"id":"productList"})  #아이템 리스트부분 추출
    lis = ul.findAll("li", {"class":"baby-product renew-badge "}) #각 아이템 추출
    
    for item in lis:
        #url
        a = item.find("a", {"class": "baby-product-link"})
        url = a.get('href')
        #print("url:", url)

        #name
        div_name = item.find("div", {"class":"name"})
        name = div_name.getText()
        #print("name:", name)

        #image
        dt_image = item.find("dt", {"class":"image"})
        image = dt_image.find("img").get('src')
        #print("image:", image)

        #price
        price = item.find("strong", {"class":"price-value"}).getText()
        #print("price:", price)
        
        row = [url, name, image, price]
        rows.append(row)
        #print(df)
    
    #print(len(lis))
    return rows
```

```python
url = "https://www.coupang.com/np/categories/497135"

pageString = getPageString(url)

df = pd.DataFrame(getProducts(pageString), columns = ['url','name','image','price'])
```
