## 동적 페이지 크롤링 
- 리뷰는 동적 페이지로 구성되어 있다
- 동적페이지란 버튼 누르면 화면은 변하는데 url은 변하지 않는 것을 의미한다
- 단순히 requests 로 얻은 html 로는 파싱 불가능 
- 동적 페이지를 구성하는 .js 파일을 찾아서 크롤링 시도
- 참고 : https://luminitworld.tistory.com/88

#### 시도 결과 
- .js 형태로 제공해주나, 리뷰 데이터가 포함되어있지 않음
- 문제 해결 못함

- Request URL: https://www.coupang.com/resources/20220120144246/np/js/lib/pacifik/messages.ko_KR.js

- Request Method: GET

- Referer: https://www.coupang.com/vp/products/4322481134?itemId=5033261361&vendorItemId=72343055403&sourceType=CATEGORY&categoryId=497035

- User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36

```python
import pandas as pd
import numpy as np
import requests # 크롤링에 사용하는 패키지
from bs4 import BeautifulSoup # html 변환에 사용함 
import json
import time
```

```python
# 리뷰 데이터 가져오기 
url = ' https://www.coupang.com/resources/20220120144246/np/js/lib/pacifik/messages.ko_KR.js'

# 헤더정보 필요
header = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36',
    'referer' : 'https://www.coupang.com/vp/products/4322481134?'
}

response = requests.get(url, headers=header)
if response.status_code == requests.codes.ok:
    print('접속성공')

response.text
```
