```python
# headers 없이 불러오면 무반응임 = 사이트에서 거부당했음을 의미 = 서버에서 봇으로 인지함
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}

res = requests.get(url, headers = headers)
```

```python
print(res) # response 가 200이면 정상 응답되었음을 의미
```

```python
res_content = res.content.decode('utf-8') # content 가져오기 
```

```python
print(res_content)
```
